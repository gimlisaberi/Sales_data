```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\emers\\OneDrive\\Desktop\\sales.csv")
df.head()
```

```python
# checking for duplicates! 

no_duplicate_df = df.drop_duplicates()
no_duplicate_df
```

```python
# Unit price must be whole number, positive!

positive_unitprice = no_duplicate_df[no_duplicate_df['UnitPrice'] > 0]
positive_unitprice
```

```python
#positive_unitprice['CustomerId'].isnull().value_counts()

cleaned_df = positive_unitprice.ffill()
cleaned_df
```

```python
# report of top 5 customers with highest no of cancelings 

canceled_invoices = cleaned_df[cleaned_df['InvoiceNumber'].str.contains('C')]
canceled_invoices.head(5)
```

```python
# removing canceled orders

final_cleaned_df = cleaned_df[~cleaned_df['InvoiceNumber'].str.contains('C')]
final_cleaned_df
```

```python
# saving changes in remaining.csv

final_cleaned_df.to_csv('remaining.csv')
```
